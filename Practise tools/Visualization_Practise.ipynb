```python
## mark ticks or values . rotation, size

import pandas as pd

import numpy as np


paper = plt.figure(figsize = [30,10])

print(df["Duration"].value_counts())  ## will get count for each value

print((df["Duration"].value_counts().plot.bar()))




plt.xticks(rotation = 45,size=12)
plt.yticks(ticks= np.arange(0,80,5),size=12)



plt.xlabel("Duration",size=20)

plt.ylabel("count",size=20)
```

```python
# upload car csv file

import io

df_car = pd.read_csv(io.BytesIO(uploaded["CAR DETAILS FROM CAR DEKHO.csv"]))

print(df_car)

```

```python
df_car.info()
```

```python
## create a histogram for km_driven

plt.figure(figsize=[20,10])

import warnings

print(df_car["km_driven"].plot.hist(color="green"))


plt.xlabel("Km_Driven",size = 15)
plt.ylabel("Frequency",size=15)
plt.title("Km_Driven distribution",size = 25)


plt.xticks(ticks=np.arange(0,1000000,100000))
plt.yticks(ticks=np.arange(0,3000,250))

warnings.filterwarnings("ignore")

plt.show()
```

```python
## create KDE plot


plt.figure(figsize=[20,10])

import warnings

print(df_car["km_driven"].plot.kde(color="green"))


plt.xlabel("Km_Driven",size = 15)
plt.ylabel("Frequency",size=15)
plt.title("Km_Driven distribution",size = 25)




warnings.filterwarnings("ignore")

plt.show()
```

```python
## create line chart ## Refer INSAID notebook for Area Chart

plt.figure(figsize=[20,10])

df_car_line = df_car.set_index("year")

print(df_car_line.head())

df_car_line[df_car_line["selling_price"] > 800000] ["selling_price"].plot.line(color="green",linestyle="-")


```

```python
## create donut chart

plt.figure(figsize=[10,10])

space = np.ones(3)/50

df_car["seller_type"].value_counts().plot.pie(cmap="inferno",legend= True,startangle=0,fontsize=14,explode=space,autopct = '%3.1f%%',wedgeprops=dict(width=0.15),shadow=True)

plt.ylabel("seller_type",size=20)
```

```python
### Create scatter  plot



import warnings

print(df_car.plot.scatter(x="selling_price",y="km_driven",figsize=[20,10]))

plt.xlabel("selling_price",size = 15)
plt.ylabel("km_driven",size=15)
plt.title("Km_driven vs Selling_price",size = 20)




plt.yticks(ticks=np.arange(0,800000,20000))


warnings.filterwarnings("ignore")

plt.show()

df_car["selling_price"].describe()

```

```python
## for regression plot use sns

import seaborn as sns

sns.regplot(x="selling_price",y="km_driven",data=df_car,color="red")
```

```python
## Create  a hex plot

import warnings

print(df_car.plot.hexbin(x="selling_price",y="km_driven",figsize=[25,10],gridsize=15,color="blue"))

plt.xlabel("selling_price",size = 15)
plt.ylabel("km_driven",size=15)
plt.title("Km_driven vs Selling_price",size = 20)



warnings.filterwarnings("ignore")

plt.show()

```

```python
## create stacked group chart

plt.figure(figsize=[15,10])


print(df_car.groupby(by=["fuel"])['selling_price'].mean().plot.bar(color="red"))
print(df_car.groupby(by=["fuel"])['km_driven'].mean().plot.bar(color="green"))
```

```python

## Create Area chart

plt.figure(figsize=[15,10])

print(df_car.groupby(by=["fuel"])['selling_price'].mean().plot.area(color="red"))
print(df_car.groupby(by=["fuel"])['km_driven'].mean().plot.area(color="green"))
```

```python
df_car.info()
```

```python

## Create bi_variate line  chart

plt.figure(figsize=[15,10])

print(df_car.groupby(by=["fuel"])['selling_price'].mean().plot.line(color="red"))
print(df_car.groupby(by=["fuel"])['km_driven'].mean().plot.line(color="green"))

plt.xlabel("Fuel",size=15)
plt.ylabel("Selling_price & Km driven",size = 15)
plt.title("Selling price  Km_driven Vs Fuel",size=20)

```

```python
## Pair plot

plt.figure(figsize=[15,10])

print(sns.pairplot(data=df_car,size = 2.5))
```

```python
## Multi variate plots

# Multi-variate Scatter plot


plt.figure(figsize=[15,10])
sns.scatterplot(x='km_driven', y='selling_price', hue='owner', style="seller_type",data=df_car)



plt.legend(fontsize=12)
```

```python

# Multi-variate Scatter plot


plt.figure(figsize=[25,10])
sns.boxplot(x='owner', y='selling_price',palette="rainbow", hue='fuel',data=df_car)





plt.legend(fontsize=12)
```

```python
## heat map


plt.figure(figsize=[25,10])

sns.heatmap(data=df_car.corr(), annot=True, cmap='viridis')
```
